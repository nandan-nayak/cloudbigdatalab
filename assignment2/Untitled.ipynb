```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
data=pd.read_csv("Assignment-1.csv")
```

```python
data
```

# Read Total profit of all months and show it using a bar plot

```python
total_profit=data['total_profit']
```

```python
total_profit
```

```python
barplot=plt.bar(data["month_number"],data["total_profit"])
plt.xticks(range(len(data)+1))
plt.xlabel("month_count")
plt.ylabel('total_profit')

plt.show()
```

# Get Total profit of all months and show line plot with the following Style properties
### Generated line plot must include following Style properties: –
#### Line Style dotted and Line-color should be red
#### Show legend at the lower right location.
#### X label name = Month Number
#### Y label name = Sold units number
#### Add a circle marker.
#### Line marker color as read
#### Line width should be 3


```python
lineplot=plt.plot(data["month_number"],data["total_profit"],color="red",marker="o")
plt.xticks(range(len(data)+1))
plt.xlabel("Month Number")
plt.ylabel('Sold units number')
plt.legend(loc=4)
mpl.rcParams['lines.linestyle'] = 'dotted'
mpl.rcParams['lines.linewidth'] = 3

plt.show()
```

# Read all product sales data and show it using a multiline plot

```python
mpl.rcParams['lines.linestyle'] = 'solid'
for i  in data:
    if i!="month_number" and i!="total_profit":
        plt.plot(data["month_number"],data[i])
plt.xticks(range(len(data)+1))
plt.legend(loc=2)
```

# Read toothpaste sales data of each month and show it using a line plot

```python
plt.plot(data["month_number"],tooth_paste)
plt.xlabel("month count")
plt.ylabel("tooth paste sales")
plt.xticks(range(len(data)+1))
plt.plot()
```

# Read face cream and facewash product sales data and show it using the multiline plot

```python
plt.plot(data["month_number"],data["facewash"])
plt.plot(data["month_number"],data["facecream"])
plt.xlabel("month number")
plt.ylabel("sales")
plt.xticks(range(len(data)+1))
plt.legend(loc=2)
plt.plot()
```

# Read sales data of bathing soap of all months and show it using a scatter plot. Save this plot to your hard disk

```python
plt.scatter(data["month_number"],data["bathingsoap"])
plt.xlabel("month number")
plt.ylabel("sales")
plt.xticks(range(len(data)+1))
plt.legend(loc=2)
plt.plot()
plt.savefig("bathingsoap.png")
```

# Read Bathing soap facewash of all months and display it using the Subplot

```python
bathingsoap=data["bathingsoap"]
```

```python
fig = plt.figure(figsize=(15, 6))
grid = plt.GridSpec(5, 30, hspace=1, wspace=1.0)
plot1= fig.add_subplot(grid[:5, :14])

plot1.bar(data["month_number"],data["bathingsoap"]);
plt.xticks(range(len(data)+1))
plt.title("bathing soap")
plt.xlabel('month number')
plt.ylabel("sales")
plot2= fig.add_subplot(grid[:5, 15:30])

plot2.plot(data["month_number"],data["facewash"])
plt.xticks(range(len(data)+1))
plt.title("face wash")
plt.xlabel('month number')
plt.ylabel("sales")
```

# Read all product sales data and show it using the stack plot

```python


plt.stackplot(data["month_number"],data["bathingsoap"])
plt.plot()


plt.stackplot(data["month_number"],data["toothpaste"])
plt.plot()


plt.stackplot(data["month_number"],data["facewash"])
plt.plot()

plt.stackplot(data["month_number"],data["shampoo"])
plt.plot()


plt.stackplot(data["month_number"],data["facecream"])
plt.plot()


plt.stackplot(data["month_number"],data["moisturizer"])
plt.plot()




plt.legend(labels=["bathingsoap","toothpaste","facewash","shampoo","facecream","mositurizer"])
plt.xlabel("month number")
plt.ylabel("sales")
plt.xticks(range(len(data)+1))
```

## Calculate total sale data for last year for each product and show it using a Pie chart


```python


labels = [ 'Moisturizer','FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'FaceCream',]
salesData   = [data ['facecream'].sum(), data ['facewash'].sum(), data ['toothpaste'].sum(),data ['bathingsoap'].sum(), data ['shampoo'].sum(), data ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%1.1f%%')


plt.title('Sales data')
plt.show()
```
